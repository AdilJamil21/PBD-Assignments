# QUESTION 2

```python
plt.pyplot.pie(df.loc[df['diagnosis'] == 'lung cancer']['state'].value_counts(), labels = ['malignant'], autopct = '%1.1f%%')
df.loc[df['diagnosis'] == 'lung cancer']['state'].value_counts()
```

```python
for x in list(df['diagnosis'].unique()):
#To Create New PieChart everytime and not to overwrite the previous. 
    plt.pyplot.figure()
    plt.pyplot.pie(df.loc[df['diagnosis'] == x]['state'].value_counts(), labels = list(df.loc[df['diagnosis'] == x]['state'].value_counts().keys()), autopct = '%1.1f%%')

#To show which type of disease is in which state we will use legend command.  
    plt.pyplot.legend(title= x)
    print(df.loc[df['diagnosis'] == x]['state'].value_counts())
```
